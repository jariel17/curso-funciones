```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-10, 10, num=N)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

y= g(x)

f_o_g = f(g(x))

plt.plot(x, f_o_g)
```

### La milla extra

```python
def l(x):
  return np.log2(x)

def e(x):
  return np.exp(x)

def va(x):
    y1 = []
    for n in x:
        if n < 0:
            y1.append(-n)
        else:
            y1.append(n)
    return y1

def c(x):
  return np.cos(x)

l_o_e = l(e(x))
plt.style.use('seaborn-v0_8-pastel')
plt.plot(x, l_o_e)
plt.title('Composicion de la funcion logaritmo con la funcion exponencial')
plt.xlabel('x')
plt.ylabel('y')
```

```python
va_o_c=va(c(x))
plt.style.use('seaborn-v0_8-poster')
plt.plot(x, va_o_c)
plt.title('Composicion de la funcion valor absoluto con coseno de x')
plt.xlabel('x')
plt.ylabel('y')
```
