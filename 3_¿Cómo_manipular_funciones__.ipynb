## Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2;

c = 5

x = np.linspace(-10,10, num=N)

y1 = f(x) + c
y2 = f(x) - c
y3 = f(x-c)
y4 = f(x+c)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
plt.style.use('seaborn-v0_8-notebook')
ax1.grid(True)
ax1.set_title('x^2 con desplazamiento hacia arriba')
ax1.plot(x,y1, label='$x^2 + 5$', color='b')
ax1.axhline(y=0, color='r', alpha=0.3)
ax1.axvline(x=0, color='r', alpha=0.3)
ax1.legend()

ax2.grid(True)
ax2.set_title('x^2 con desplazamiento hacia abajo')
ax2.plot(x,y2, label='$x^2 - 5$', color='c')
ax2.axhline(y=0, color='r', alpha=0.3)
ax2.axvline(x=0, color='r', alpha=0.3)
ax2.legend()

ax3.grid(True)
ax3.set_title('x^2 con desplazamiento hacia la derecha')
ax3.plot(x,y3, label='$(x-5)^2$', color='salmon'), 
ax3.axhline(y=0, color='r', alpha=0.3)
ax3.axvline(x=0, color='r', alpha=0.3)
ax3.legend()

ax4.grid(True)
ax4.set_title('x^2 con desplazamiento hacia la izquierda')
ax4.plot(x,y4, label='$(x+5)^2$', color='orange')
ax4.axhline(y=0, color='r', alpha=0.3)
ax4.axvline(x=0, color='r', alpha=0.3)
ax4.legend()
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontalmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontalmente en un factor de $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y1 = c*f(x)
y2 = (1/c)*f(x)
y3 = f(c*x)
y4= f(1/c*x)


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
plt.style.use('seaborn-v0_8-notebook')
ax1.grid(True)
ax1.set_title('sin(x) con alargamiento vertical')
ax1.plot(x,y1, label='$2sin(x)$', color='b')
ax1.axhline(y=0, color='r', alpha=0.3)
ax1.axvline(x=0, color='r', alpha=0.3)
ax1.legend()

ax2.grid(True)
ax2.set_title('sin(x) con compresion vertical')
ax2.plot(x,y2, label='$1/2(sin(x))$', color='c')
ax2.axhline(y=0, color='r', alpha=0.3)
ax2.axvline(x=0, color='r', alpha=0.3)
ax2.legend()

ax3.grid(True)
ax3.set_title('sin(x) con compresion horizontal')
ax3.plot(x,y3, label='$sin(2x)$', color='salmon'), 
ax3.axhline(y=0, color='r', alpha=0.3)
ax3.axvline(x=0, color='r', alpha=0.3)
ax3.legend()

ax4.grid(True)
ax4.set_title('sin(x) con alargamiento horizontal')
ax4.plot(x,y4, label='$sin((1/2)*x)$', color='orange')
ax4.axhline(y=0, color='r', alpha=0.3)
ax4.axvline(x=0, color='r', alpha=0.3)
ax4.legend()
```

## Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(x)
y2 = f(-x)

fig, axes = plt.subplots(nrows=1, ncols=2)
plt.style.use('seaborn-v0_8-deep')
axes[0].grid(True)
axes[0].set_title('Grafica de x^3')
axes[0].plot(x,y)
axes[0].axhline(y=0, color='r', alpha=0.5)
axes[0].axvline(x=0, color='r', alpha=0.5)

axes[1].grid(True)
axes[1].set_title('Grafica de x^3 reflejada')
axes[1].plot(x,y2)
axes[1].axhline(y=0, color='r', alpha=0.5)
axes[1].axvline(x=0, color='r', alpha=0.5)
```
