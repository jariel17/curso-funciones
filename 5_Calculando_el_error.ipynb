# Dependencias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([1.2, 2, 3.2, 2.5, 5, 6, 4, 8])
y = np.array([2, 3, 3.4, 3.1, 4, 4.7, 3.8, 7])
n = len(x)
```

Lets get the indices to sort x

```python
indices = np.argsort(x)
x = x[indices]
y = y[indices]
```

```python
print(x)
print(y)
```

```python
plt.scatter(x, y, marker='x', color='red', s=30)
plt.grid(True)
plt.show()
```

Ahora graficaré mi función de predicción (elegida aleatoriamente para este ejemplo)

```python
N = 8

m = 0.9

b = 1

def f(x):
  return m*x+b

x1 = np.linspace(1,8, num=N)

y1 = f(x)

fig, ax = plt.subplots()
ax.plot(x1,y1, 'bo:')
ax.grid(True)
```
